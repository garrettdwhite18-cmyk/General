```python
import pandas as pd
from google.colab import files

uploaded = files.upload()

filename = list(uploaded.keys())[0]
FILEDATA = filename
SHEETNAME = 'spot'
spot = (
    pd.read_excel(filename, sheet_name=SHEETNAME)
    .rename(columns={'Unnamed: 0': 'field'})
    .set_index('field')
)

display(spot)
```

```python
EXPRY_CHAIN_LOC = 1
SHEETNAME = pd.ExcelFile(FILEDATA).sheet_names[EXPRY_CHAIN_LOC]
options = pd.read_excel(FILEDATA, sheet_name=SHEETNAME).set_index('ticker')
display(options.loc[options['option type'].str.contains('C')])
display(options.loc[options['option type'].str.contains('P')])
```

```python
print(options.columns)
```

```python
import numpy as np
from scipy.stats import norm

F = spot.loc['price'].values[0]

calls = options[options['option type'] == 'C'].copy()

atm = calls.iloc[(calls['strike price'] - F).abs().argsort()].iloc[0]

K = atm['strike price']
r = atm['finance rate']
T = atm['days to expiration'] / 365
sigma = 0.0076
market_price = atm['price']

d1 = (np.log(F / K) + 0.5 * sigma**2 * T) / (sigma * np.sqrt(T))
d2 = d1 - sigma * np.sqrt(T)

model_price = np.exp(-r * T) * (F * norm.cdf(d1) - K * norm.cdf(d2))

print(f"ATM strike: {K}")
print(f"Implied vol (futures price): {sigma:.6f}")
print(f"T (years): {T:.4f}")
print(f"Finance rate: {r:.6f}")
print(f"Model price: {model_price:.6f}")
print(f"Market price: {market_price}")
print(f"Difference: {model_price - market_price:.6f}")
```

```python
import numpy as np
import pandas as pd
from scipy.stats import norm
import matplotlib.pyplot as plt

F = spot.loc['price'].values[0]

T = options['days to expiration'].iloc[0] / 365

r = options['finance rate'].iloc[0]

strikes = []
market_prices = []
black_prices = []
option_types = []

for idx, row in options.iterrows():
    K = row['strike price']
    option_type = row['option type']
    market_price = row['price']

    sigma = row['implied vol'] / 25

    d1 = (np.log(F / K) + 0.5 * sigma**2 * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)

    if option_type == 'C':
        black = np.exp(-r*T) * (F * norm.cdf(d1) - K * norm.cdf(d2))
    else:  # Put
        black = np.exp(-r*T) * (K * norm.cdf(-d2) - F * norm.cdf(-d1))

    strikes.append(K)
    market_prices.append(market_price)
    black_prices.append(black)
    option_types.append(option_type)

df_plot = pd.DataFrame({
    'Strike': strikes,
    'Market Price': market_prices,
    'Black Price': black_prices,
    'Option Type': option_types
})

plt.figure(figsize=(10,6))
calls = df_plot[df_plot['Option Type'] == 'C']
plt.plot(calls['Strike'], calls['Market Price'], 'bo-', label='Calls Market')
plt.plot(calls['Strike'], calls['Black Price'], 'b--', label='Calls Blackâ€“76')

puts = df_plot[df_plot['Option Type'] == 'P']
plt.plot(puts['Strike'], puts['Market Price'], 'ro-', label='Puts Market')
plt.plot(puts['Strike'], puts['Black Price'], 'r--', label='Puts Blackâ€“76')

plt.xlabel('Strike Price')
plt.ylabel('Option Price')
plt.title('Nearest Expiry Option Chain: Blackâ€“76 vs Market')
plt.legend()
plt.grid(True)
plt.show()

df_plot['Diff'] = df_plot['Black Price'] - df_plot['Market Price']
display(df_plot[['Strike','Option Type','Market Price','Black Price','Diff']])
```

Blackâ€“76 values can differ from market prices for several reasons. Market prices reflect actual bid-ask spreads, rounding, and microstructure effects, whereas Blackâ€“76 produces a theoretical mid-price. Additionally, the model assumes a constant interest rate and uses a single implied volatility, while the market incorporates volatility skews, smiles, and small adjustments for funding costs or cash flows in the underlying futures. These factors together can create small but noticeable differences between model and market prices.

The American exercise feature of these options is generally not very significant. For interest rate futures options, early exercise rarely adds value because futures are marked-to-market daily and holding an option to expiration usually does not carry additional risk. Only deep in-the-money options or scenarios with special cash-flow considerations might justify early exercise, so using Blackâ€“76, which assumes European-style exercise, is typically a very good approximation. But it does generally add some level of value to be able to exercise early for certain situations.

To compare this to historical data, we can calculate the 60-day rolling volatility of the SOFR rate using daily rate data. This gives a sense of how realized volatility compares to the implied vol quoted by the market.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from google.colab import files
uploaded = files.upload()

ref_rates = pd.read_excel('ref_rates.xlsx', sheet_name='data')

display(ref_rates.head())
print(ref_rates.columns)

ref_rates['SOFR'] = pd.to_numeric(ref_rates['SOFR'], errors='coerce')

sofr_series = ref_rates['SOFR'].dropna()

sofr_diff = sofr_series.diff()

sofr_vol_60d = sofr_diff.rolling(window=60).std()

sofr_vol_60d_annualized = sofr_vol_60d * np.sqrt(252)

sofr_vol_df = pd.DataFrame({
    'SOFR': sofr_series,
    'SOFR_diff': sofr_diff,
    '60d_vol_annualized': sofr_vol_60d_annualized
})

display(sofr_vol_df.tail())

plt.figure(figsize=(10,5))
plt.plot(sofr_vol_df.index, sofr_vol_df['60d_vol_annualized'], label='60-Day Rolling Vol (Annualized)')
plt.title('60-Day Rolling Volatility of SOFR')
plt.xlabel('Date Index')
plt.ylabel('Annualized Volatility')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt


F = spot.loc['price'].values[0]

T = options['days to expiration'].iloc[0] / 365

r = options['finance rate'].iloc[0]

black_prices = []
futures_vol = []

for idx, row in options.iterrows():
    K = row['strike price']
    option_type = row['option type']

    sigma = row['implied vol'] / 25
    futures_vol.append(sigma)

    d1 = (np.log(F / K) + 0.5 * sigma**2 * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)

    if option_type == 'C':
        price = np.exp(-r*T)*(F*norm.cdf(d1) - K*norm.cdf(d2))
    else:
        price = np.exp(-r*T)*(K*norm.cdf(-d2) - F*norm.cdf(-d1))

    black_prices.append(price)

options['Black_Price'] = black_prices
options['Futures_Vol'] = futures_vol


options['Notional_OI'] = options['strike price'] * options['option contract size'] * options['open int']
total_notional_oi = options['Notional_OI'].sum()
print(f"Total Notional Open Interest: {total_notional_oi:,.2f}")

calls = options[options['option type'] == 'C']
plt.figure(figsize=(10,5))
plt.plot(calls['strike price'], calls['price'], 'bo-', label='Market Price')
plt.plot(calls['strike price'], calls['Black_Price'], 'b--', label='Blackâ€“76 Price')
plt.xlabel('Strike Price')
plt.ylabel('Call Option Price')
plt.title('Call Options: Market vs Blackâ€“76 Model')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10,5))
plt.plot(options['strike price'], options['implied vol'], 'ro-', label='Implied Vol')
plt.xlabel('Strike Price')
plt.ylabel('Implied Volatility')
plt.title('Implied Volatility Across Strikes')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

calls = options[options['option type'] == 'C']
puts = options[options['option type'] == 'P']

plt.figure(figsize=(10,5))
plt.plot(calls['strike price'], calls['open int'], 'b-o', label='Calls')
plt.plot(puts['strike price'], puts['open int'], 'r-o', label='Puts')

plt.xlabel('Strike Price')
plt.ylabel('Open Interest (Contracts)')
plt.title('Open Interest Across Strikes: Calls vs Puts')
plt.legend()
plt.grid(True)
plt.show()
```
