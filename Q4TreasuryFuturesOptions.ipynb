##Part A) a, v, p for Swaption and TY during April 2025

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

beta = 0.5
T = 1.0

vol_df = pd.read_excel(file_name, sheet_name="black_vol_pct")
strike_df = pd.read_excel(file_name, sheet_name="absolute_strikes_pct")
fwd_df = pd.read_excel(file_name, sheet_name="atm_forward")

vol_df['date'] = pd.to_datetime(vol_df['date'])
strike_df['date'] = pd.to_datetime(strike_df['date'])
fwd_df['date'] = pd.to_datetime(fwd_df['date'])

# Convert to decimals
vol_df.iloc[:,1:] /= 100
strike_df.iloc[:,1:] /= 100
fwd_df['atm_fwd_pct'] /= 100

strike_cols = vol_df.columns[1:]

def sabr_vol(F, K, T, alpha, beta, rho, nu):

    if F == K:
        FK = F**(1-beta)
        term1 = alpha / FK
        term2 = 1 + (
            ((1-beta)**2 / 24)*(alpha**2/FK**2)
            + (rho*beta*nu*alpha)/(4*FK)
            + ((2-3*rho**2)/24)*nu**2
        ) * T
        return term1 * term2

    logFK = np.log(F/K)
    FK = (F*K)**((1-beta)/2)

    z = (nu/alpha) * FK * logFK
    xz = np.log((np.sqrt(1-2*rho*z+z*z)+z-rho)/(1-rho))

    term1 = alpha / (FK * (1 + ((1-beta)**2/24)*logFK**2 + ((1-beta)**4/1920)*logFK**4))
    term2 = z/xz

    term3 = 1 + (
        ((1-beta)**2/24)*(alpha**2/FK**2)
        + (rho*beta*nu*alpha)/(4*FK)
        + ((2-3*rho**2)/24)*nu**2
    ) * T

    return term1 * term2 * term3


def loss(params, F, K, vols):

    alpha, rho, nu = params

    if alpha <= 0 or nu <= 0 or abs(rho) >= 1:
        return 1e6

    model = np.array([sabr_vol(F,k,T,alpha,beta,rho,nu) for k in K])
    return np.sum((model - vols)**2)


results = []

for i in range(len(vol_df)):

    F = fwd_df.loc[i,'atm_fwd_pct']

    vols = vol_df.loc[i,strike_cols].values
    K = strike_df.loc[i,strike_cols].values

    x0 = [0.02, 0.0, 0.5]

    res = minimize(loss, x0, args=(F,K,vols), method='Nelder-Mead')

    alpha, rho, nu = res.x

    results.append({
        "date": vol_df.loc[i,'date'],
        "alpha": alpha,
        "rho": rho,
        "nu": nu
    })

sabr_df = pd.DataFrame(results)
sabr_df.set_index("date", inplace=True)

print(sabr_df.head())
```

##a, v, p for TY

```python
import scipy.optimize
from scipy.stats import norm
from tqdm import tqdm

file_name = "ty_option_surfaces_p7.xlsx"
df_m2025 = pd.read_excel(file_name, sheet_name="CBT TY TY M2025")
df_m2025['date'] = pd.to_datetime(df_m2025['date'])

def sabr_x(f_nought, strike, beta):
    return (f_nought * strike) ** ((1 - beta) / 2)

def sabr_y(f_nought, strike, beta):
    return (1 - beta) * np.log(f_nought / strike)

def sabr_a(f_nought, strike, beta, alpha):
    x = sabr_x(f_nought, strike, beta)
    y = sabr_y(f_nought, strike, beta)
    return alpha / (x * (1 + (y**2 / 24) + ((y**4) / 1920)))

def sabr_b(f_nought, strike, beta, alpha, rho, nu, ttm):
    x = sabr_x(f_nought, strike, beta)
    return 1 + ((((1 - beta) ** 2 / 24) * (alpha ** 2 / (x ** 2))) + ((rho * beta * nu * alpha) / (4 * x)) + ((nu ** 2 * (2 - 3 * rho ** 2)) / 24)) * ttm

def sabr_phi(f_nought, strike, alpha, beta, nu):
    x = sabr_x(f_nought, strike, beta)
    return (nu / alpha) * x * np.log(f_nought / strike)

def sabr_chi(f_nought, strike, alpha, beta, rho, nu):
    phi = sabr_phi(f_nought, strike, alpha, beta, nu)
    return np.log((np.sqrt(1 - 2 * rho * phi + phi**2) + phi - rho) / (1 - rho))

def sabr_implied_vol(f_nought, strike, ttm, alpha, beta, rho, nu):
    if f_nought == strike:
        return alpha / (f_nought ** (1 - beta))
    a = sabr_a(f_nought, strike, beta, alpha)
    b = sabr_b(f_nought, strike, beta, alpha, rho, nu, ttm)
    phi = sabr_phi(f_nought, strike, alpha, beta, nu)
    chi = sabr_chi(f_nought, strike, alpha, beta, rho, nu)
    return (a * b * phi) / chi

def delta_to_strike(F, delta, sigma, T, is_call=True):
    phi = 1 if is_call else -1
    return F * np.exp(-phi * norm.ppf(phi * delta) * np.sqrt(T) * sigma + 0.5 * sigma**2 * T)

def fit_alpha_to_atm(atm_vol, f_nought, beta, nu, rho, ttm):
    def objective(alpha):
        b = sabr_b(f_nought, f_nought, beta, alpha, rho, nu, ttm)
        model_atm_vol = alpha * b / (f_nought ** (1 - beta))
        return (model_atm_vol - atm_vol) ** 2
    result = scipy.optimize.minimize(objective, [0.02], bounds=[(0.001, None)])
    return result.x[0]

def calibrate_sabr_atm(f_nought, strikes, ttm, market_vols, beta, atm_vol):
    rho_init = 0.7
    nu_init = 0.2
    def sabr_atm_objective(params):
        rho, nu = params
        alpha_derived = fit_alpha_to_atm(atm_vol, f_nought, beta, nu, rho, ttm)
        model_vols = [sabr_implied_vol(f_nought, strike, ttm, alpha_derived, beta, rho, nu) for strike in strikes]
        return np.sum((model_vols - market_vols) ** 2)
    result = scipy.optimize.minimize(sabr_atm_objective, [rho_init, nu_init],
                                     bounds=[(-0.9,0.9),(0.001,None)],
                                     method='Nelder-Mead', tol=1e-11)
    return result.x  # rho, nu

def rowfit_sabr_atm(row, beta=0.5):
    future_price = row['Future Price']
    option_ttm = row['Expiration Option']
    atm_vol = row['P50dvol']
    put_col_names = [c for c in row.index if c.startswith('P') and c.endswith('dvol')]
    put_vols = row[put_col_names].values.astype(float)
    put_deltas = np.array([int(c[1:-4])/100 for c in put_col_names])
    put_strikes = np.array([delta_to_strike(future_price, -d, v, option_ttm, is_call=False)
                            for d,v in zip(put_deltas, put_vols)])
    fitted_rho, fitted_nu = calibrate_sabr_atm(future_price, put_strikes, option_ttm, put_vols, beta, atm_vol)
    fitted_alpha = fit_alpha_to_atm(atm_vol, future_price, beta, fitted_nu, fitted_rho, option_ttm)
    return fitted_alpha, fitted_nu, fitted_rho

df_april = df_m2025[(df_m2025['date'] >= '2025-04-01') & (df_m2025['date'] <= '2025-04-30')].reset_index(drop=True)

alpha_list, nu_list, rho_list, dates = [], [], [], []

for idx, row in tqdm(df_april.iterrows(), total=len(df_april), desc="Calibrating M2025 SABR"):
    alpha, nu, rho = rowfit_sabr_atm(row, beta=0.5)
    alpha_list.append(alpha)
    nu_list.append(nu)
    rho_list.append(rho)
    dates.append(row['date'])

sabr_m2025_april = pd.DataFrame({
    'date': dates,
    'alpha': alpha_list,
    'nu': nu_list,
    'rho': rho_list
})
sabr_m2025_april.set_index('date', inplace=True)
print(sabr_m2025_april.head())

plt.figure(figsize=(14,5))
plt.plot(sabr_m2025_april.index, sabr_m2025_april['alpha'], marker='o')
plt.title("M2025 SABR α (April 2025, Puts Only)")
plt.xlabel("Date")
plt.ylabel("α")
plt.grid(True)
plt.show()

plt.figure(figsize=(14,5))
plt.plot(sabr_m2025_april.index, sabr_m2025_april['nu'], marker='o', color='orange')
plt.title("M2025 SABR ν (April 2025, Puts Only)")
plt.xlabel("Date")
plt.ylabel("ν")
plt.grid(True)
plt.show()

plt.figure(figsize=(14,5))
plt.plot(sabr_m2025_april.index, sabr_m2025_april['rho'], marker='o', color='green')
plt.title("M2025 SABR ρ (April 2025, Puts Only)")
plt.xlabel("Date")
plt.ylabel("ρ")
plt.grid(True)
plt.show()
```

For TY M2025 options during April 2025, the SABR parameters exhibit much higher alpha (~0.7–0.8), mostly positive or near-zero rho (~0–0.9), and moderate nu (~0.4–2.5). The large alpha reflects the higher absolute price level of the underlying Treasury futures (~110), producing substantial ATM volatility over the short-dated option horizon. The positive or near-zero rho often generates a slightly upward-skewed or symmetric smile, indicating that calls are slightly more expensive than puts on some days. The moderate nu captures short-term fluctuations in volatility, but because the options are short-dated, vol-of-vol is less pronounced than for longer-dated swaptions. Overall, the TY option smile is sharper in the wings and sometimes tilted upward, in contrast to the flatter, negatively skewed swaption smile, reflecting differences in expiry, underlying levels, and market pricing of tail risk.

To analyze the largest parameter moves in TY M2025 options during April 2025, we first computed day-over-day absolute changes in the SABR parameters ν and ρ. We then summed these changes to create a single movement criterion (|Δν| + |Δρ|) for each day. Using this, we identified the top 5 days with the largest TY movements.

For comparison, we extracted the corresponding day-over-day changes in ν and ρ for the 1Yx5Y swaptions on the same dates. This allows us to assess not just the magnitude of movements but also the direction of change between the two markets, highlighting differences in how short-dated futures options (TY) and longer-dated swaptions respond to market volatility.

```python
ty_delta_nu = ty_april['nu'].diff()
ty_delta_rho = ty_april['rho'].diff()

swpt_delta_nu = swpt_april['nu'].diff()
swpt_delta_rho = swpt_april['rho'].diff()

ty_criteria = ty_delta_nu.abs() + ty_delta_rho.abs()
top5_dates = ty_criteria.nlargest(5).index

comparison_df = pd.DataFrame({
    'TY ΔNu': ty_delta_nu.loc[top5_dates],
    'Swpt ΔNu': swpt_delta_nu.loc[top5_dates],
    'TY ΔRho': ty_delta_rho.loc[top5_dates],
    'Swpt ΔRho': swpt_delta_rho.loc[top5_dates],
})

comparison_df['Nu Direction'] = np.sign(comparison_df['TY ΔNu'] * comparison_df['Swpt ΔNu'])
comparison_df['Rho Direction'] = np.sign(comparison_df['TY ΔRho'] * comparison_df['Swpt ΔRho'])

comparison_df['Nu Magnitude Ratio'] = comparison_df['TY ΔNu'].abs() / comparison_df['Swpt ΔNu'].abs()
comparison_df['Rho Magnitude Ratio'] = comparison_df['TY ΔRho'].abs() / comparison_df['Swpt ΔRho'].abs()

display(comparison_df)
```

Overall, they moved together more often than not, which makes sense.

Short-dated futures options (TY) are far more sensitive to daily volatility shocks than longer-dated swaptions, reflecting both shorter expiry and higher gamma exposure. The swaption market moves much more slowly and exhibits smaller adjustments, consistent with its longer tenor and smoother risk profile Overall, these results confirm that short-dated interest rate options capture tail risk more acutely, while swaptions behave in a more muted and smoothed fashion over the same period.

The negative tilt and symmetry of the swaption smile also suggest the market assigns slightly more weight to downward rate moves, but overall the long tenor smooths out extreme tail risk.

##Part D) Possible Reasons for any Differences and Scaling

Reasons for Differences of Smiles:

Several factors likely contribute to the differences observed between TY options and 1Y×5Y swaptions. First, the option tenor is critical: TY options are short-dated (often weeks to a few months), so even small absolute moves in interest rates produce large relative changes, amplifying the wings of the smile. Swaptions, in contrast, have a longer effective horizon, which naturally smooths out short-term fluctuations and reduces the apparent tail risk.

Second, moneyness and strike coverage differs: the TY option surface we analyzed primarily covers strikes within ±5–10% of the forward, focusing on near-the-money and moderately OTM options. Swaptions span a wider range, up to ±75% in delta, capturing more extreme strikes. This difference in strike sampling affects the apparent steepness of the wings when plotted on the same scale.

Third, market microstructure and liquidity can play a role. Short-dated futures options (TY) are highly liquid and sensitive to daily trading flows, which can exaggerate spikes in implied volatility for out-of-the-money strikes. Swaptions, being longer-dated and less frequently traded at extreme strikes, reflect smoother expectations of long-term rate distributions rather than immediate market pressures.

Finally, skew expectations and risk premiums differ. The negative skew observed in swaptions reflects market anticipation that rates are more likely to decline than spike upward over the long horizon, while TY options show a more symmetric or upward-skewed profile due to short-term uncertainty and higher sensitivity to immediate tail events.
